# WORLD BANK DATA

```python
wb_indicators = {
    "population_total": "SP.POP.TOTL",
    "population_65_UP": "SP.POP.65UP.TO",
    "population_15_64": "SP.POP.1564.TO",
    "densite": "EN.POP.DNST",
    "pib_total": "NY.GDP.MKTP.CD",
    "pib_par_habitant": "NY.GDP.PCAP.CD",
    "croissance_pib": "NY.GDP.MKTP.KD.ZG",
    "participation_femmes": "SL.TLF.CACT.FE.ZS",
    "densite_urbaine": "EN.URB.LCTY",
    "population_urbaine": "SP.URB.TOTL.IN.ZS"

}
```

```python
import requests
import time

def get_wdi_indicator(code, name):
    records = []
    page = 1
    
    while True:
        url = (f"https://api.worldbank.org/v2/country/all/indicator/"
               f"{code}?format=json&per_page=20000&page={page}&date=1896:2016")

        response = requests.get(url)
        
        try:
            data = response.json()
        except:
            print(f" Erreur JSON → {name}")
            break

        if len(data) < 2 or data[1] is None:
            print(f" Aucun résultat pour {name} (page {page})")
            break

        for d in data[1]:
            if d["value"] is not None:
                records.append({
                    "country": d["country"]["value"],
                    "year": int(d["date"]),
                    name: d["value"]
                })

        total_pages = data[0].get("pages", 1)
        if page >= total_pages:
            break

        page += 1
        time.sleep(0.05)
    
    return pd.DataFrame(records)
```

#### Télecharger tous les indicateurs dans des dataframes stockés dans une liste

```python
#  DOWNLOAD DE TOUS LES INDICATEURS
dataFrames = []

for name, code in wb_indicators.items():
    df_indicators = get_wdi_indicator(code, name)
    if not df_indicators.empty:
        dataFrames.append(df_indicators)
        print(f"Télechargement de : {len(df_indicators)} lignes pour {name}")
    else:
        print(f"Vide → {name}")
```

Chaque indicateur n’a pas les mêmes pays / années. Si l’année/pays existe dans au moins un DataFrame → on garde la ligne et si une colonne manque → NaN

```python
# Fusion progressive
df_wb = dataFrames[0]
for d in dataFrames[1:]:
    df_wb = df_wb.merge(d, on=["country","year"], how="outer")
```

```python
df_wb.shape
```

```python
df_wb.isnull().sum()
```

## dataframe final

```python
df_wb_final = df_wb.merge(
    df_noc,
    left_on=["country"],
    right_on=["region"],
    how="inner"
)
df_wb_final.head()
```

```python
df_wb_final.shape
```

```python
df_wb_final.isnull().sum()
```

#### supprimer la colonne region car c'est la meme que country

```python
df_wb_final.drop(columns=["region"], inplace=True)
```

#### Traitement des valeurs null

même si par exemple PIB de certains pays diminuer avec les années, l’interpolation reste parfaitement correcte car l’interpolation respecte la tendance réelle, même si elle est négative

ce code remplit uniquement les valeurs manquantes (NaN) quand il y a des valeurs avant et après

```python
df_wb_final = df_wb_final.sort_values(by=["country", "year"])

# Colonnes économiques à interpoler
econ_cols = ["pib_total", "pib_par_habitant", "croissance_pib","population_urbaine","densite_urbaine","densite"]

# Interpolation par pays
for col in econ_cols:
    df_wb_final[col] = df_wb_final.groupby("country")[col].transform(lambda x: x.interpolate(method="linear"))
```

L’interpolation est utile uniquement quand :

les données représentent une évolution naturelle dans le temps
